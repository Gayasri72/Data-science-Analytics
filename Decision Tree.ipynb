```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
raw_data=pd.read_csv("zoo.csv")
```

```python
raw_data.head()
```

```python
raw_data.shape
```

```python
x=raw_data.iloc[:,1:17]
```

```python
x.shape
```

```python
x.head()
```

```python
y=raw_data.iloc[:,17]
```

```python
y.shape
```

```python
y.head()
```

### Splitting the data

```python
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Building the classifier

```python
zoo_classifier = DecisionTreeClassifier(random_state=0)
```

```python
zoo_classifier.fit(X_train,y_train)
```

### Testing the model

```python
zoo_classifier.score(X_test,y_test)
```

### Making prediction

```python
zoo_classifier.predict(X_test[10:15])
```
